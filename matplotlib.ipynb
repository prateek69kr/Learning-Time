# Pie Chart 📊🌒

```python
#to display one digit decimal
```

```python
categories = ["Freshmen", "Sophomores", "Juniors", "Seniors"]
values = np.array([330, 250, 265, 225])

plt.pie(values, labels = categories,
                autopct= "%1.1f%%")
```

```python
#cokor set for each slice
```

```python
categories = ["Freshmen", "Sophomores", "Juniors", "Seniors"]
values = np.array([330, 250, 265, 25])
color = ["pink", "blue", "green", "red"]
plt.pie(values, labels = categories,
                autopct= "%1.1f%%",
               colors=color)

plt.title("Institute Of Stealology",
            fontsize =15)
```

```python
#to highjlight we use explode
```

```python
categories = ["Freshmen", "Sophomores", "Juniors", "Seniors"]
values = np.array([330, 250, 265, 25])
plt.pie(values, labels = categories,
                autopct= "%1.1f%%",
              explode= [0, 0.1, 0, 0],
               shadow = True)

plt.title("Institute Of Stealology",
            fontsize =15)
```

```python
#to rotate startangle
```

```python
categories = ["Freshmen", "Sophomores", "Juniors", "Seniors"]
values = np.array([330, 250, 265, 25])
plt.pie(values, labels = categories,
                autopct= "%1.1f%%",
              explode= [0, 0.1, 0, 0],
               shadow = True,
                startangle=180)

plt.title("Institute Of Stealology",
            fontsize =15)
```

# Histogram 📊

>Histogram A visual representation of the distribution of quantitative data
They group values into bins (intervals)
and counts how many fall in each range.



```python
scores = np.random.normal(loc=80, scale = 10, size = 100)
plt.hist(scores)
```

```python
#to stop exceedind limit we use clip function
```

```python
scores = np.random.normal(loc=80, scale = 10, size = 100)
scores = np.clip(scores, 0, 100)

plt.hist(scores)
```

```python
#add some customization : bins(interval or number of histogram)
```

```python
scores = np.random.normal(loc=80, scale = 10, size = 100)
scores = np.clip(scores, 0, 100)

plt.hist(scores, bins=10,
         color = "lightgreen",
         edgecolor = "black")
plt.title("Test Scores",
          fontsize = 15)
```
